```python
# get the env variables values
token = os.getenv("TOKEN")
token_type = os.getenv("TOKEN_TYPE")
print(token)
signin_response = {
    "access_token": token,
    "token_type": token_type
}
my_headers  = {
    'Authorization' : f'Bearer ' + signin_response.get("access_token")
}
print(my_headers)
```

```python
url = "https://www.opensourcenepal.com/api/imports"

# set params
params = {
    "limit": 100,
    "offset": 0
}

response = requests.get(url,
                        params=params,
                        headers=my_headers)

# check if the requests was successful (status code 200)
if response.status_code == 200:
    # extract the data from the response
    data = response.json()
    # print the data 
    print(data)
    print(type(data))
else:
    print("Error: Failed to retrieve data from the API.")
```

```python
country_count = {} # country: count

for country in top_import_countries:
    if country in country_count:
        # we've seen this country --> so increment it
        country_count[country] += 1
    else:
        # we have not seen this country --> add to our map
        country_count[country] = 1

# find most common country and its count
most_common_country = max(country_count, key=country_count.get)
# in above line: we are finding the key (country) in the country_count dictionary that has the highest value (count)

count = country_count[most_common_country]
print("Most repeated country is: ", most_common_country)
print("Max no of occurrence in this dataframe is: ", count)
```

```python
## Dictionary of countries imported

# Create a dictionary where country names are the keys and their corresponding import amounts are the values.

country_imports_dict = dict(zip(gbapi_df['imported_from'], gbapi_df['amount']))
print(country_imports_dict)
```

```python
## Nested Dictionary
# Create a nested dictionary where each country has an inner dictionary containing its year-wise import data.
nested_dict = {}
for index, row in gbapi_df.iterrows():
    country = row['imported_from']
    year = row['year']
    amount = row['amount']

    if country not in nested_dict:
        nested_dict[country] = {}
    nested_dict[country][year] = amount

print(nested_dict)
```
